```python
x = df.iloc[:, 2:].values
y = df.iloc[:, 1].values
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
model = RandomForestClassifier()

params = {
    'n_estimators': [10,50,100,1000],
    'criterion': ['gini', 'entropy', 'log_loss'],
    'max_depth': [3,4,5],
    'max_features':['sqrt', 'log2'],
}

GS = GridSearchCV(estimator= model, param_grid= params, cv=5, verbose=5)
GS.fit(x, y)
```

```python
model_best = GS.best_estimator_
best_params = GS.best_params_
```

```python
important_score = model_best.feature_importances_
important_score
```

```python
df.columns
```

```python
df_scores = pd.DataFrame({"Features":df.columns[2:],"Scores":important_score})
df_scores = df_scores.reset_index()
df_scores = df_scores.sort_values(by="Scores",ascending=False)
df_scores
```

```python
# mean_scores = df_scores["Scores"].mean()
df_scores = df_scores[df_scores["Scores"] > 0.000]
df_scores
```
